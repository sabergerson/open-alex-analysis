```python
import pandas as pd

df = pd.read_csv('all_domains.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)
color = 'black'
color_map = ["#d6e635ff", "#18dec1ff", "#3f3994ff"]

index = df[df['domain']=='Semantic computing'].index[0]
year = range(2012,2022)
data1 = df.loc[index,'educ_per'].replace('[','').replace(']','').split(', ')
data1 = [float(i) for i in data1]
data2 = df.loc[index,'comp_per'].replace('[','').replace(']','').split(', ')
data2 = [float(i) for i in data2]
data3 = df.loc[index,'coll_per'].replace('[','').replace(']','').split(', ')
data3 = [float(i) for i in data3]

fig, ax1 = plt.subplots()
ax1.set_xlabel('Year', fontsize=24)
ax1.set_ylabel('Proportion of Total Publications \n', color='black', fontsize=24)
ax1.stackplot(year, data1, data2, data3, labels=['Academia','Industry','Collaboration'], colors=color_map)
ax1.tick_params(axis='both', labelsize=16)
ax1.set_ylim([0.8, 1.0])
ax1.legend(loc='upper left', prop={'size': 18})

data4 = df.loc[index,'mean'].replace('[','').replace(']','').split(', ')
data4 = [float(i) for i in data4]

ax2 = ax1.twinx()
ax2.set_ylabel('\nMean Compute Mentions', color='black', fontsize=24)  # we already handled the x-label with ax1
ax2.plot(year, data4, color='white', label='Compute Mentions', linewidth = 3, marker='o', markersize = 5)
ax2.tick_params(axis='y', labelcolor=color, labelsize=16)
ax2.set_ylim([0, 1])
ax2.legend(loc='upper right', prop={'size': 18})

fig.tight_layout()
plt.title('Publication Proportion and Compute Usage (Semantic Computing)\n', fontsize=24)
plt.gcf().set_size_inches(20, 10)
plt.show()
```
